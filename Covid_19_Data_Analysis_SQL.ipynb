SQLAlchemy is a library that facilitates the communication between Python programs and databases. 

SQLite is a C-language library that implements a small, fast, self-contained, high-reliability, full-featured, SQL database engine. SQLite is the most used database engine in the world.

echo = FALSE prevents code, but not the results from appearing in the finished file.

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite://', echo = False)

## export the dataframe as a table 'playstore' to the sqlite engine

final_data.to_sql('covid', con = engine)
```

# ANALYSING THE DATA USING SQL QUERIES 

```python
#select query

sql = ''' select * from covid''';

df_sql = pd.read_sql_query(sql, con = engine)

df_sql.head()
```

```python
#select the country available 

sql = '''select Country from covid''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(100)
```

```python
# Countries with most active cases

sql = ''' select Country, Active from covid group by Country order by Active  desc limit 15''';

df_sql = pd.read_sql_query(sql, con = engine)

df_sql.head(15)
```

```python
#countries with successful discharges (recover) 
sql = '''select Country , recover_100 from covid group by country order by recover_100 desc limit 15''';

df_sql = pd.read_sql_query(sql, con = engine)

df_sql.head(15)
```

```python
#countries with most 10 new deaths

sql = ''' select Country , new_deaths	from covid group by Country order by new_deaths	desc limit 11''';

df_sql = pd.read_sql_query(sql, con = engine)

df_sql.head(11)
```

```python
#Compute the recovered rate 

sql = '''select * from (select Country, Recovered, Confirmed, 
          round(100 * (round(Recovered,2) / round(Confirmed,2)),2) as Recovered_Ratio_Calculated
          from covid group by Country order by Deaths desc limit 10 )a

          order by a.Recovered_Ratio_Calculated''';

df_sql = pd.read_sql_query(sql,con = engine)
df_sql.head(11)
```

```python
# Max total cases vs Min Total cases 
sql = ''' select * from (select ' The Minimum number of cases : ' || Country as place, min(Confirmed) as Max_Min_Cases from covid 
          union all
          select ' The Maximum number of cases : '|| Country,  max(Confirmed) as Max_Min_Cases from covid )a

          order by a.Max_Min_Cases desc ''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Max Deaths vs Min Deaths 

sql = '''select * from (select 'The Minimum number of deaths :' || Country as place, min(Deaths) as Max_Min_Deaths from covid
        union all
        select 'The Maximum number of deaths :' || Country as place , max(Deaths) as Max_Min_Deaths from covid)a

        order by a.Max_Min_Deaths desc''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
# Max Active cases  vs Min Active cases  

sql = '''select * from (select 'The Minimum number of Active cases  :' || Country as place, min(Active) as Max_Min_Active_Cases from covid
        union all
        select 'The Maximum number of Active cases :' || Country as place , max(Active) as Max_Min_Active_Cases from covid)a

        order by a.Max_Min_Active_Cases desc''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql
```

```python
#Countries with Most new cases and new Deaths 

sql = '''select Country , n_cases , new_deaths from covid group by Country order by n_cases  desc''';

df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(21)
```
